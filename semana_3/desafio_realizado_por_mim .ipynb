### Acesse a api da camara, e recupere os reembolsos do ano de 2018

```python
import requests

url = "http://www.camara.leg.br/cotas/Ano-2018.csv.zip"

# fazendo o download do conteúdo do arquivo
r = requests.get(url)
```

```python
# abrindo o arquivo
file = open("Ano-2018.csv.zip", "wb")

# escrever o conteúdo
file.write(r.content)
file.close()
```

```python
from zipfile import ZipFile

# abrindo o .zip
zip_file = ZipFile("Ano-2018.csv.zip", 'r')
```

```python
zip_file.namelist()
# lista os arquivos que tem dentro do zip
```

```python
# extrai o arquivo
zip_file.extract(member="Ano-2018.csv", path="reembolso-2018")
```

```python
# fecha o arquivo zip
zip_file.close()
```

### Leia o Arquivo csv 

```python
import pandas as pd
```

```python
pd.read_csv("reembolso-2018/Ano-2018.csv", sep=";")
```

```python
# indica o tipo de cada uma das colunas
DTYPE = {
    'txNomeParlamentar': str,
    'ideCadastro': str,
    'nuCarteiraParlamentar': str,
    'nuLegislatura': str,
    'sgUF': str,
    'sgPartido': str,
    'codLegislatura': str,
    'numSubCota': str,
    'txtDescricao': str,
    'numEspecificacaoSubCota': str,
    'txtDescricaoEspecificacao': str,
    'txtFornecedor': str,
    'txtCNPJCPF': str,
    'txtNumero': str,
    'indTipoDocumento': str,
    'datEmissao': str,
    'vlrDocumento': float,
    'vlrGlosa': str,
    'vlrLiquido': float,
    'numMes': str,
    'numAno': str,
    'numParcela': str,
    'txtPassageiro': str,
    'txtTrecho': str,
    'numLote': str,
    'numRessarcimento': str,
    'nuDeputadoId': str,
    'ideDocumento': str,
}

#lê o arquivo csv e transforma em um dataframe, já com a leitura exata dos tipos
df_reembolsos = pd.read_csv("reembolso-2018/Ano-2018.csv", sep=";", dtype=DTYPE)
```

### Converta a coluna `datEmissao` para `datetime`, e descubra quantos reembolsos aconteram no primeiro semestre

### Limpe a coluna `txtCNPJCPF`: ela deve conter somente dígitos numéricos

```python
# vamos trocar qualquer caracter que não seja um dígito número por '', essa operação remove os dígitos
df_reembolsos["txtCNPJCPF"] = df_reembolsos["txtCNPJCPF"].str.replace(r'\D', '', regex=True)
# o str mostra que a coluna a ser alterada tem que ser lida como string
# no caso o 'r' mostra que é uma expressão regular e o 'D' indica que é qualquer valor que não número
```
