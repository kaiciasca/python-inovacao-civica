```python
#testes da aula 1.7 - importar várias tabelas de uma vez usando o glob
```

```python
import glob
```

```python
dados_todos_estados = glob.glob('/home/kaiciasca/Documentos/Kaian/Jornalismo_de_Dados/curso-python/python-inovacao-civica/modulo_1/conjunto_dados/consulta_*.csv')
print(dados_todos_estados)
```

```python
lista_df = []

for dados_estado in dados_todos_estados:
    df_estado = pd.read_csv(dados_estado, sep=';', encoding='latin_1')
    lista_df.append(df_estado)
```

```python
print(lista_df[0:1])
```

```python
df_completo = pd.concat(lista_df, axis=0, ignore_index=True)
df_completo.shape
```
