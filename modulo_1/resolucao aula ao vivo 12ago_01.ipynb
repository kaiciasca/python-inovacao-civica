```python
# como ler vários arquivos csv e converter de uma só vez para dataframe Pandas
```

```python
import pandas as pd
import glob
```

```python
lista_arquivos = glob.glob('conjunto_dados/consulta*.csv')
print(lista_arquivos)
```

```python
lista_df = []
```

```python
for csv in lista_arquivos:
    df = pd.read_csv(csv, sep=';', encoding='latin_1')
    lista_df.append(df)
```

```python
print(lista_df)
```

```python
candidaturas_PE_RN = pd.concat(lista_df, axis=0, ignore_index=True)
```

```python
print(candidaturas_PE_RN)
```
